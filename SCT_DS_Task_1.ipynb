```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_267553.csv', skiprows=4)

latest_year = '2024'

top10 = df[['Country Name', latest_year]].dropna().sort_values(
    by=latest_year, ascending=False
).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['Country Name'], top10[latest_year])
plt.title('Top 10 Most Populous Countries')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
